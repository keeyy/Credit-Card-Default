## Import Library

Import library that we need

```python
import numpy as np
import pandas as pd
import joblib
```

Import the model that we saved

```python
# Load all files
with open('model_nb.pkl', 'rb') as file_1:
  model_nb= joblib.load(file_1)

with open('model_scaler.pkl', 'rb') as file_2:
  model_scaler=joblib.load(file_2)

with open('list_num_col.txt', 'rb') as file_3:
  num_col= joblib.load(file_3)

with open('list_cat_col.txt', 'rb') as file_4:
  cat_col= joblib.load(file_4)
```

Now make a random data from generate random

```python
# Generate random data
random = np.random.default_rng(seed=2895)
limit_balance = random.choice(np.arange(10000, 800000, 10000), size=30)
pay_0 = random.integers(-0.2, 8.0, size=30)
pay_2 = random.integers(-0.2, 7.0, size=30)
pay_3 = random.integers(-0.2, 7.0, size=30)
pay_4 = random.integers(-0.2, 8.0, size=30)
pay_5 = random.integers(-0.2, 7.0, size=30)
pay_6 = random.integers(-0.2, 7.0, size=30)
data_inf = pd.DataFrame({'limit_balance' : limit_balance, 'pay_0' : pay_0, 'pay_2' : pay_2, 'pay_3' : pay_3, 'pay_4' : pay_4, 'pay_5' : pay_5, 'pay_6' : pay_6})
data_inf
```

## Split

Split the numerical and categorical columns

```python
# Split the data
data_inf_num = data_inf[num_col]
data_inf_cat = data_inf[cat_col]
```

## Scale and Encode

Scale the numerical column

```python
# Scale the numerical
data_inf_scale = model_scaler.transform(data_inf_num)
```

We don't need to encode the categorical because the value already in the form of numbers

## Concate

Concate the numerical that already scaled and the categorical column

```python
# Concate the data
data_inf_final = np.concatenate([data_inf_scale,data_inf_cat], axis=1)
```

## Predict

Now predict the data

```python
# Predict
y_pred_inf= model_nb.predict(data_inf_final)
y_pred_inf
```

All predicted as default payment next month

```python
data_inf.insert(7, 'Predict', y_pred_inf)
```
